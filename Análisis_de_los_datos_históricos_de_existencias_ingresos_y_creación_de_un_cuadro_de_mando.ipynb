Pregunta 1 - Extracción de datos de acciones de Tesla utilizando yfinance - 2 Puntos

```python
import yfinance as yf

# Descargar datos históricos de Tesla
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(start='2020-01-01', end='2025-03-22')

# Mostrar las primeras filas del DataFrame
print(tesla_data.head())
```

Pregunta 2 - Extracción de datos de ingresos de Tesla utilizando Webscraping - 1 Puntos

```python
import requests
from bs4 import BeautifulSoup

# URL de la página de relaciones con inversores de Tesla
url = 'https://ir.tesla.com/'

# Realizar la solicitud HTTP
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar el contenido del atributo title
title_content = soup.title.string
print(title_content)
```

Pregunta 3 - Extracción de datos de acciones de GameStop utilizando yfinance - 2 Puntos

```python
# Descargar datos históricos de GameStop
gamestop = yf.Ticker('GME')
gamestop_data = gamestop.history(start='2020-01-01', end='2025-03-22')

# Mostrar las primeras filas del DataFrame
print(gamestop_data.head())
```

Pregunta 4 - Extracción de datos de ingresos de GameStop utilizando Webscraping - 1 Puntos

```python
import yfinance as yf
import pandas as pd

# Descargar datos históricos de GameStop
gme = yf.Ticker('GME')
gme_data = gme.history(period='1y')  # Datos del último año

# Mostrar las primeras filas del DataFrame
print(gme_data.head())
```

Pregunta 5 - Tablero de acciones e ingresos de Tesla - 2 Puntos

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los datos de precios e ingresos
df_tesla = pd.DataFrame({
    'Fecha': tesla_data.index,
    'Precio': tesla_data['Close'],
    'Ingresos': [17450] * len(tesla_data)  # Ingresos constantes como ejemplo
})

# Graficar los datos
plt.figure(figsize=(12, 6))
plt.plot(df_tesla['Fecha'], df_tesla['Precio'], label='Precio de Acción')
plt.plot(df_tesla['Fecha'], df_tesla['Ingresos'], label='Ingresos', linestyle='--')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Comparación de Precio de Acción e Ingresos de Tesla')
plt.legend()
plt.show()
```
